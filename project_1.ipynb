SALES ANALYSIS - PHYTHON PROJECT (USING NUMPY,PANDAS,MATPLOTLIB,SEABORN)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/phython.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.isnull().sum()
```

```python
df['quantity'].fillna(df['quantity'].median(),inplace=True)
```

```python
df.info()
```

```python
df['price'] = df.groupby('category')['price'].transform(lambda x : x.fillna(x.mean()))
```

```python
df.info()
```

```python
df['payment_mode'].fillna(df['payment_mode'].mode()[0], inplace = True)
```

```python
df.info()
```

```python
df['revenue']=df['quantity']*df['price']
```

```python
df['revenue']
```

```python
df
```

```python
total_rev = df['revenue'].sum()
```

```python
total_rev
```

```python
df['month'] = df['order_date'].dt.month
monthly_sales = df.groupby('month')['revenue'].sum()
```

```python
monthly_sales
```

```python
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
sns.barplot(x='category',y='revenue',data=df)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
top_customers = df.groupby('customer_name')['revenue'].sum().sort_values(ascending = False)
top_customers.plot(kind='bar')
plt.title("total customer by revenue")
plt.show()
```

```python
df.info()
```

```python
df['cost_price'] = df['price']*0.65
```

```python
df['cost_price']
```

```python
df['profit'] = (df['price'] - df['cost_price'])*df['quantity']
```

```python
df['profit']
```

```python
profit_category = df.groupby('category')['price'].sum()
profit_category
```

```python
profit_category.plot(kind='bar')
plt.title("Profit by Category")
plt.ylabel("Profit")
plt.show()
```

```python
customer_orders = df.groupby('customer_id')['order_id'].count()
customer_orders
```

```python
retained_customers = customer_orders[customer_orders > 1].count()
total_customers = customer_orders.count()
retained_rate = (retained_customers / total_customers) * 100
retained_rate
```

```python
churn_customers = customer_orders[customer_orders== 1].count()
churn_rate = (churn_customers/total_customers)*100
churn_rate
```

```python
aov = df['revenue'].sum()/df['order_id'].nunique()
aov
```

```python
df['revenue_pct'] = (df['revenue']/df['revenue'].sum())*100
df[['product','revenue_pct']]
```

```python
kpi_summary = {
    "total_revenue" : df['revenue'].sum(),
    "total_profit" : df['order_id'].sum(),
    "retained_customers %":retained_rate,
    "churn_customers %": churn_rate,
    "average_order_value" : aov
}
pd.DataFrame(kpi_summary, index=['KPI'])
```
